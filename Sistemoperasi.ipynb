```python
import matplotlib.pyplot as plt
import networkx as nx

# Buat directed graph
G = nx.DiGraph()

# Daftar proses dan resource
processes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
resources = ['R', 'S', 'T', 'U', 'V', 'W']

# Tambahkan node dengan tipe berbeda
G.add_nodes_from(processes, type='process')
G.add_nodes_from(resources, type='resource')

# Tambahkan edges berdasarkan deskripsi
edges = [
    ('R', 'A'),  # A holds R
    ('A', 'S'),  # A wants S
    ('B', 'T'),  # B wants T
    ('C', 'S'),  # C wants S
    ('U', 'D'),  # D holds U
    ('D', 'S'),  # D wants S
    ('D', 'T'),  # D wants T
    ('T', 'E'),  # E holds T
    ('E', 'V'),  # E wants V
    ('W', 'F'),  # F holds W
    ('F', 'S'),  # F wants S
    ('V', 'G'),  # G holds V
    ('G', 'U'),  # G wants U
]

G.add_edges_from(edges)

# Gaya visualisasi
pos = nx.spring_layout(G, seed=42)
node_colors = ['lightblue' if G.nodes[n]['type'] == 'process' else 'lightgreen' for n in G.nodes]
node_shapes = {'process': 'o', 'resource': 's'}

# Gambar node proses dan resource secara terpisah
fig, ax = plt.subplots(figsize=(12, 8))

for ntype in node_shapes:
    nodelist = [n for n in G.nodes if G.nodes[n]['type'] == ntype]
    nx.draw_networkx_nodes(G, pos, nodelist=nodelist, node_shape=node_shapes[ntype],
                           node_color=['lightblue' if ntype == 'process' else 'lightgreen'],
                           node_size=1200, ax=ax)

# Draw edges and labels
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=20, edge_color='gray', width=2, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', ax=ax)

plt.title("Resource Allocation Graph (RAG)", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```
