```python
#bulan - pengeluaran perminggu
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
bulan = ['jan','feb','mar','apr','mai','juni','juli','agt','sep','oct','nov','dec']

Minggu1 = np.random.randint(200000,500001,12)
Minggu2 = np.random.randint(200000,500001,12)
Minggu3 = np.random.randint(200000,500001,12)
Minggu4 = np.random.randint(200000,500001,12)

data = {
    'Bulan' : bulan,
    'Minggu_1' : Minggu1,
    'Minggu_2' : Minggu2,
    'Minggu_3' : Minggu3,
    'Minggu_4' : Minggu4,
}

df = pd.DataFrame(data)
df['Total_Pengeluaran'] = df[['Minggu_1','Minggu_2','Minggu_3','Minggu_4']].sum(axis=1)
df
```

```python
#diagram bar

plt.figure(figsize=(12,6))
plt.bar(df['Bulan'],df['Total_Pengeluaran'],color='black')
plt.title("Pengeluaran perbulan")
plt.xticks(rotation=-45)
plt.xlabel("Bulan")
plt.ylabel("Rupiah")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))


for minggu in ['Minggu_1','Minggu_2','Minggu_3','Minggu_4']:
    plt.plot(df['Bulan'],df[minggu],marker='o', label=minggu)

plt.title('Pengeluaran per minggu per bulan')
plt.xlabel("Bulan")
plt.ylabel("rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation= 20)
plt.yticks(rotation= -20)
plt.tight_layout()
plt.show()
```

```python
#diagram pie

plt.figure(figsize=(8,8))
plt.title("Proporsi Pengeluaran perbulan")
plt.pie(df["Total_Pengeluaran"], labels=df['Bulan'], startangle=90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

```python
#regrresi liniar
#AI,ML 
from sklearn.linear_model import LinearRegression as LR

df['Bulan_Angka'] = np.arange(1,13)

x = df[['Bulan_Angka']]
y = df['Total_Pengeluaran']

model = LR()
model.fit(x,y)

prediksi = model.predict([[13]])[0]

print(f"Pengeluaran di bulan 13: {prediksi:,.0f}")
```
